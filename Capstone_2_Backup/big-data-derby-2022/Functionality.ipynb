```python
entries=pd.read_csv('nyra_start_table_new.csv')
races=pd.read_csv('nyra_race_table.csv')
tracking=pd.read_csv('nyra_tracking_table.csv')
tracking.program_number=tracking.program_number.str.replace(' ','-')
```

```python
def get_run(race_date,race_id,horse_id):
    run=tracking[(tracking.race_date==race_date)
           & (tracking.program_number==horse_id) 
           & (tracking.race_number==race_id)]
    run=run.sort_values('trakus_index')
    run['gps']=run.apply(lambda r:(r.latitude,r.longitude),axis=1)
    track=run.track_id.values[0]
    return track,run[['trakus_index','gps']]
def get_race(race_date,race_id):
    runs=tracking[(tracking.race_date==race_date) 
            & (tracking.race_number==race_id)].copy()
    runs['gps']=runs.apply(lambda r:(r.latitude,r.longitude),axis=1)
    runs=runs[['program_number','trakus_index','gps']].sort_values('trakus_index')
    return runs.groupby('program_number')
```

```python
def get_race_info(date,race):
    info=races[(races.race_date==date) &
               (races.race_number==race)
              ][['distance_id','course_type','track_condition','run_up_distance','post_time']]
    info=info.to_dict()
    horses=entries[(entries.race_date==date) & 
                   (entries.race_number==race)].program_number.values.tolist()
    finishes=entries[(entries.race_date==date) & 
                     (entries.race_number==race)].position_at_finish.values.tolist()
    return info,horses,finishes
```

```python
def dist(pt_0,pt_1):
    return geopy.distance.geodesic(pt_0,pt_1)
def analyze_run(run):
    start=run[run.trakus_index==1].iloc[0]
    stats=pd.DataFrame()
    stats['time (s)']=(run['trakus_index']-1)*0.25
    stats['gps']=run.gps
    stats['gps_prior']=stats.gps.shift(1)
    stats=stats[stats['time (s)']>0]
    stats['delta (ft)']=stats.apply(lambda x: dist(x.gps_prior,x.gps).ft,axis=1)
    stats['delta (mi)']=stats.apply(lambda x: dist(x.gps_prior,x.gps).mi,axis=1)
    stats.drop('gps_prior',axis=1,inplace=True)
    stats['velo (mph)']=stats['delta (mi)']*60*60/0.25
    stats['accel']=stats['velo (mph)']/0.25
    stats['jerk']=stats['accel (mph)']/0.25
    stats['dist. (ft)']=stats['delta (ft)'].cumsum()
    stats['dist. (mi)']=stats['delta (mi)'].cumsum()
    stats['latitude']=stats.gps.apply(lambda x:x[0])
    stats['longitude']=stats.gps.apply(lambda x:x[1])
    stats.drop('gps',axis=1,inplace=True)
    return stats
```

```python
def F(x,a,b,c):
    y=a*np.square(x)
    y=y+(b*x)
    y=y+c
    return y
```
